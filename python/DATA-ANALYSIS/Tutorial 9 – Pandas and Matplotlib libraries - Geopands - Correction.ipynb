# Exercise 2 - Seaborn

Seaborn is a Python data visualization library based on matplotlib.

Using Seaborn official site (http://seaborn.pydata.org/), or tutorials such
« Pandas & Seaborn - A guide to handle & visualize data in Python »
(https://tryolabs.com/blog/2017/03/16/pandas-seaborn-a-guide-to-handle-visualize-dataelegantly/),
answer the following questions.

1. Create dataframe df from Pokemon.csv (Moodle) with Name column as index.
2. Display bar charts for count of ‘Type 1’ for each generation of pokemons.
3. Display box plots of ‘Total’ according to generations. Repeat but with swarm plots over the box plots.
4. Display split violin plots of ‘Attack’ according to ‘Type 1’ for Legendary and non legendary pokemons.
5. Remove columns ‘#’, ‘Sp. Atk’ and ‘Sp. Def’ from df. Create dataframe df2 as df without ‘Total’, ‘Generation’ and ‘Legendary’ columns. Display pair plot for each variable in df2.
6. Display as heatmap of correlation coefficients between variables of df2.
7. Display factor plot of ‘Attack’ according to ‘Generation’ for each type of Pokemon (‘Type 1’ variable).

```python
import seaborn as sb
```

```python
# 1. Create dataframe df from Pokemon.csv (Moodle) with Name column as index.
df = pd.read_csv('pokemon.csv', index_col='Name')
```

```python
# 2. Display bar charts for count of ‘Type 1’ for each generation of pokemons.
ax = sb.countplot(x='Generation' , hue='Type 1', data=df)
ax.figure.set_size_inches((14, 6))
ax.legend(ncol=3, title= 'Type 1')
```

```python
# 3. Display box plots of ‘Total’ according to generations. Repeat but with swarm plots over the box plots.
ax = sb.boxplot(x='Generation', y='Total', data=df, color='0.5');
ax.figure.set_size_inches((14, 6))

ax = sb.boxplot(x='Generation', y='Total', data=df, color='0.5'); sb.swarmplot(x='Generation', y='Total', data=df, color='0.2', alpha=0.8)
ax.figure.set_size_inches((14, 6))
```

```python
# 4. Display split violin plots of ‘Attack’ according to ‘Type 1’ for Legendary and non legendary pokemons.
ax = sb.violinplot(x="Type 1", y="Attack", data=df, hue="Legendary", split=True, inner='quart')
ax.figure.set_size_inches((14, 6))
```

```python
# 5. Remove columns ‘#’, ‘Sp. Atk’ and ‘Sp. Def’ from df. Create dataframe df2 as df without ‘Total’, ‘Generation’ and ‘Legendary’ columns. Display pair plot for each variable in df2.
df.drop(['#'] + [col for col in df.columns if col.startswith('Sp.')], axis=1, inplace=True)
df2 = df.drop(['Total','Generation','Legendary'], axis=1)
sb.pairplot(df2, markers='.', kind='reg')
```

```python
# 6. Display as heatmap of correlation coefficients between variables of df2.
ax = sb.heatmap(df2.corr(), annot=True,cmap='RdBu_r', center=0, cbar_kws={'label':'Correlation coefficient'})
ax.set_aspect('equal')
```

```python
# 7. Display factor plot of ‘Attack’ according to ‘Generation’ for each type of Pokemon (‘Type 1’ variable).
sb.factorplot(x='Generation', y='Attack', data=df,hue='Type 1', col='Type 1', col_wrap=3, kind='swarm');
```

```python
# import geopandas and folium
import geopandas as gp
import folium
import pandas as pd
```

# Import data with spatial informations

```python
df = pd.read_csv('gis/pokemon-geo.csv')
df.info()
df
```

# Convert pandas df to a geopandas df

```python
#convert a pandas dataframe into a geopandas dataframe
gdf = gp.GeoDataFrame(
    df, geometry=gp.points_from_xy(df.x, df.y), crs="EPSG:4326"
)

#if needed, import matplotlib
#geopandas + matplotlib will produce a static map
import matplotlib
gdf.plot()
```

```python
gdf
```

Add a folium map (based on leaflet js) and visualise

```python
#Create a folium map - dynamic and interactive (but slow)
map = folium.Map(
    location=[0, 0],
    tiles="cartodbpositron",
    zoom_start=2,
)

#Try changing the background to
#tiles="Stamen Terrain"
# or tiles="OpenStreetMap" or tiles="Stamen Toner"

map
```

```python
# Create a geometry list from the GeoDataFrame
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in gdf.geometry]

geo_df_list_low = geo_df_list[:100]

# Iterate through list and add a marker for each volcano, color-coded by its type.
i = 0
for coordinates in geo_df_list:
    # assign a color marker for the type of volcano, Strato being the most common
    # if gdf.Name[i] == "M":
    #     type_color = "green"
    # elif gdf.HP[i] == "W":
    #     type_color = "blue"
    # else:
    #     type_color = "purple"

    type_color = "blue"

    # Place the markers with the popup labels and data
    map.add_child(
        folium.Marker(
            location=coordinates,
            popup="Name: "
            + str(gdf['Name'][i])
            + "<br>"
            + "Gender: "
            + str(gdf['HP'][i])
            + "<br>"
            + "Attack: "
            + str(gdf['Attack'][i]),
            icon=folium.Icon(color="%s" % type_color),
        )
    )
    i = i + 1
```
